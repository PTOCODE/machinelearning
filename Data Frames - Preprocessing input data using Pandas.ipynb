```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv("data/szegedWeatherHistory.csv")
data.dtypes
```

```python
x1=data['Temperature (C)']
x2=data['Humidity']
y=data['Apparent Temperature (C)']
print(x1)
print(x2)
print(y)
```

```python
from sklearn.model_selection import train_test_split
train_set,test_set=train_test_split(data,test_size=0.2,random_state=42)
print("TRAIN DATA SIZE %d, TEST DATA SIZE %d"%(len(train_set),len(test_set)))
print("Train data set is:",train_set)
print("Test data set is:",test_set)
```

```python
x1=train_set['Temperature (C)']
x2=train_set['Humidity']
y=train_set['Apparent Temperature (C)']
print(x1)
print(x2)
print(y)
```

```python
x1=np.array(x1).reshape(len(train_set),1)
x2=np.array(x2).reshape(len(train_set),1)
y=np.array(y).reshape(len(train_set),1)
c1=np.array([1 for i in range(len(train_set))]).reshape(len(train_set),1)
X=np.concatenate([c1,x1,x2],axis=1)
print(X)
```

```python
from numpy.linalg import lstsq
THETA_LS=lstsq(X,y)[0]
print(THETA_LS)
```

```python
xx1=test_set['Temperature (C)']
xx2=test_set['Humidity']
y_test=test_set['Apparent Temperature (C)']
print(x1)
print(x2)
print(y)
```

```python
xx1=np.array(xx1).reshape(len(test_set),1)
xx2=np.array(xx2).reshape(len(test_set),1)
y_test=np.array(y_test).reshape(len(test_set),1)
c2=np.array([1 for i in range(len(test_set))]).reshape(len(test_set),1)
X_test=np.concatenate([c2,xx1,xx2],axis=1)
print(X_test)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(X,y)
y_predict=model.predict(X_test)
print(y_predict)
```

```python
print(model.score(X_test,y_test))
```
