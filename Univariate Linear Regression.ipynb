```python
import numpy as np
from sklearn.linear_model import LinearRegression 
```

```python
X= [6,8,10,14,18]#List of  Independent variable for training
y=[7,9,13,17.5,18]#List of Dependent variable for training
```

```python
X=np.array(X).reshape(5,1)#Transforming Independent variable into a vector(1 column) 
y=np.array(y).reshape(5,1)#Transforming Dependent variable into a 1-D vector
```

```python
print(X)
print(y)
```

```python
model =LinearRegression() #Creating an object named model of class Linear Regression
model.fit(X,y)
```

```python
print(model.intercept_)#Printing intercept i.e distance from y-axis; theta_0 in equation y = theta_0 + theta_1 * x
print(model.coef_)#Printing coefficient or slope; theta_1 in above equation
```

```python
X_test=[8,9,11,16,12]#List of  Independent variable for testing
y_test=[11,8.5,15,18,11]#List of Dependent variable for testing
X_test=np.array(X_test).reshape(5,1) #Tranforming into 1-D vector
y_test=np.array(y_test).reshape(5,1)
print(X_test)
print(y_test)
y_pred=model.predict(X_test)#predicting the y values for given X values by calling the method predict 
```

```python
from matplotlib import pyplot as plt
print("Plot of Training data")
plt.title("Pizza Price Plotted against diameter")
plt.plot(X_test,y_pred) #Plotting the predicted y value versus X value (Straight Line)
plt.plot(X_test,y_test,'k.') #Plotting the actual test vlaue of y versus X test value
plt.axis([0,25,0,25])
plt.plot()


```

```python
from matplotlib import pyplot as plt
print("Plot of Training data")
plt.title("Pizza Price Plotted against diameter")
plt.plot(X_test,y_pred)
plt.plot(X_test,y_test,'k.')
plt.axis([0,25,0,25])
for i in range(0,5):#plotting the distance between actual value and predicted value 
    plt.plot([X_test[i],X_test[i]],[y_pred[i],y_test[i]],color="red",linewidth="1")
```
